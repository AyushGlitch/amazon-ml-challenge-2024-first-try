```python
import pandas as pd
import numpy as np                                          
```

```python
df= pd.read_csv("../tested.csv")
print(df.shape)
df= df.drop(['entity_name', 'extracted_text'], axis=1)
df.head()
```

```python
df['entity_value_num'] = df['entity_value'].str.rsplit(' ', n=1).str[0]
df['entity_value_unit'] = df['entity_value'].str.rsplit(' ', n=1).str[1]

df['predicted']= df['predicted'].fillna("Val Unit")
df['predicted_value_num'] = df['predicted'].str.rsplit(' ', n=1).str[0]
df['predicted_value_unit'] = df['predicted'].str.rsplit(' ', n=1).str[1]

print(df.shape)
df.head()
```

```python
df['entity_value_num'] = pd.to_numeric(df['entity_value_num'], errors='coerce')
df.dropna(subset=['entity_value_num'], inplace=True)

df['predicted_value_num'] = pd.to_numeric(df['predicted_value_num'], errors='coerce')
df['predicted_value_num'] = df['predicted_value_num'].fillna(0)

print(df.shape)
df.head()
```

```python
units_list= ['fluid ounce', 'milligram', 'millilitre', 'decilitre', 'inch', 'yard', 'microlitre', 'cup', 'litre', 'cubic inch', 'microgram', 'quart', 'volt', 'watt', 'kilovolt', 'millimetre', 'pint', 'kilowatt', 'cubic foot', 'ton', 'pound', 'ounce', 'gallon', 'centimetre', 'kilogram', 'imperial gallon', 'millivolt', 'centilitre', 'gram', 'metre', 'foot']

df = df[df['entity_value_unit'].isin(units_list)]
print(df.shape)
```

```python
# Convert numeric columns to strings before concatenation
df['cleaned_entity_value'] = df['entity_value_num'].astype(str) + ' ' + df['entity_value_unit']
df['cleaned_predicted_value'] = df['predicted_value_num'].astype(str) + ' ' + df['predicted_value_unit']

# Display the shape and the first 60 rows of the DataFrame
print(df.shape)
df.head()
```

```python
df2= df[['cleaned_entity_value', 'cleaned_predicted_value']]
print(df2.shape)
df2.head()
```

```python
df2.to_csv("../clean_tested.csv")
```
